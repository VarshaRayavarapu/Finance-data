```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Finance_data.csv' , index_col = None)
df.head()
```

```python
lst = ['What are your savings objectives?', 
       'Reason_Equity',
       'Reason_Bonds',
       'Reason_Mutual', 
       'Reason_FD', 
       'Avenue' ]
```

```python
len(lst)
```

```python
print(df.columns.tolist())
```

```python
for i in range(1, len(lst)+1):
    plt.tight_layout()
    df[lst[i-1]].value_counts().plot(kind='bar')
    plt.title(lst[i-1])
    plt.show()
```

```python
df['Reason_FD'].value_counts().plot(kind='bar')
```

```python
sns.swarmplot(x = df['gender'] ,  y = df['age'], data=df, palette='husl')
```

### 🐝 Swarm Plot Explanation
This plot visualizes the age distribution of respondents, separated by gender. Each dot represents a person. It's useful to see age range and clustering patterns.

```python
sns.barplot(x=df['gender'], y=df['Government_Bonds'], data=df, palette='husl')
```

### 📊 Bar Plot Explanation
This chart shows the average rank given to Government Bonds, separated by gender. Lower rank means higher preference. It includes a title and proper labels.

```python
df.describe()
```

```python
sns.lmplot(x ='age', y='Mutual_Funds' , data=df)
```

### 📈 Linear Regression Plot
This visualization explores the relationship between age and preference for mutual funds using a regression line.

```python
# Plot average investment ranks for female respondents
df[df['gender'] == 'Female'][['Mutual_Funds', 'Equity_Market', 'Debentures',
                              'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'age']].mean().plot(
    kind='bar', color='red', title='Female Average Investment Preferences'
)

plt.ylabel('Average Rank (Lower = More Preferred)')
plt.xticks(rotation=45)
plt.show()
```

```python
df[df['gender'] == 'Male'][['Mutual_Funds', 'Equity_Market', 'Debentures',
                            'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'age']].mean().plot(
    kind='bar', color='green', title='Male Average Investment Preferences'
)

plt.ylabel('Average Rank (Lower = More Preferred)')
plt.xticks(rotation=45)
plt.show()
```
